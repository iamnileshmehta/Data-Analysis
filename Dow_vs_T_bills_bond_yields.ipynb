```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Fetch historical data
# Dow Jones (^DJI), 10-Year Treasury Yield (^TNX)
dow = yf.download('^DJI', start='2004-01-01', end='2024-11-22')
bond_yield = yf.download('^TNX', start='2004-01-01', end='2024-11-22')
```

```python
# Normalize Dow for comparison
dow['Normalized'] = (dow['Close'] / dow['Close'].iloc[0]) * 100

# Bond yields: Use actual yield percentage
bond_yield['Yield (%)'] = bond_yield['Close']

# Plotting
plt.figure(figsize=(12, 8))

# Plot Dow (normalized)
plt.plot(dow.index, dow['Normalized'], label='Dow Jones (Normalized)', color='green')

# Plot Bond Yields
plt.plot(bond_yield.index, bond_yield['Yield (%)'], label='10-Year Treasury Yield (%)', color='red')

# Highlight significant events
plt.axvline(pd.Timestamp('2008-09-15'), color='purple', linestyle='--', label='2008 Crisis')
plt.axvline(pd.Timestamp('2020-03-01'), color='orange', linestyle='--', label='COVID-19 Crash')

# Add labels, title, legend, and grid
plt.title('Dow Jones vs 10-Year U.S. Treasury Yield (2004-2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Performance / Yield (%)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch historical data for 3-Month T-Bill (^IRX)
t_bill = yf.download('^IRX', start='2004-01-01', end='2024-01-01')

# Use the 'Close' column to represent the yield
t_bill['Yield (%)'] = t_bill['Close']

# Plotting the T-Bill Yield
plt.figure(figsize=(12, 6))

plt.plot(t_bill.index, t_bill['Yield (%)'], label='3-Month T-Bill Yield (%)', color='blue')

# Highlight significant events
plt.axvline(pd.Timestamp('2008-09-15'), color='purple', linestyle='--', label='2008 Crisis')
plt.axvline(pd.Timestamp('2020-03-01'), color='orange', linestyle='--', label='COVID-19 Crash')

# Add labels, title, legend, and grid
plt.title('3-Month T-Bill Yield (2004-2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Yield (%)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
