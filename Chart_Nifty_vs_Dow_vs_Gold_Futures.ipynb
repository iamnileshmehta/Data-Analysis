```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
nifty = yf.download("^NSEI", start= "2004-01-01", end= "2024-11-22")
dow = yf.download('^DJI', start='2004-01-01', end='2024-11-22')
gold  = yf.download('GC=F', start='2004-01-01', end='2024-11-22')
```

```python
# Normalize the 'Close' prices
nifty['Normalized'] = (nifty['Close'] / nifty['Close'].iloc[0]) * 100
dow['Normalized'] = (dow['Close'] / dow['Close'].iloc[0]) * 100
gold['Normalized'] = (gold['Close'] / gold['Close'].iloc[0]) * 100
```

```python


# Plotting
plt.figure(figsize=(12, 8))

plt.plot(nifty.index, nifty['Normalized'], label='Nifty 50', color='blue')
plt.plot(dow.index, dow['Normalized'], label='Dow Jones', color='green')
plt.plot(gold.index, gold['Normalized'], label='Gold', color='gold')

# Highlight significant events
plt.axvline(pd.Timestamp('2008-09-15'), color='red', linestyle='--', label='2008 Crisis')
plt.axvline(pd.Timestamp('2020-03-01'), color='purple', linestyle='--', label='COVID-19 Crash')

# Add labels, title, legend, and grid
plt.title('Nifty 50 vs Dow Jones vs Gold (2004-2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Normalized Value (Index = 100)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
